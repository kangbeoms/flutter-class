```r
sales <- c(210,110,400,550,600,130)
week <- c('월','화','수','목','금','토')

ratio <- sales / sum(sales) * 100
label <- paste(week,"\n",ratio,'%')
# 백그라운드 색 
par(family = 'AppleGothic',bg= 'white')
pie(sales,
    main = '영업실적 파이 차트',
    col = rainbow(length((sales))),
    labels = label
)
            
```

```r
# 3D Pie 차트
#install.packages("plotrix")
library(plotrix)
```

```r
sales <- c(210,110,400,550,600,130)
week <- c('월','화','수','목','금','토')

ratio <- sales / sum(sales) * 100
label <- paste(week,"\n",ratio,'%')
# 백그라운드 색 
par(family = 'AppleGothic',bg= 'white')
pie3D(sales,
    main = '영업실적 파이 차트',
    col = rainbow(length((sales))),
    labels = label,
    # 3D 에서 퍼센트가 높으면 위로 튀어나오게 하기
    explode = 0.05,
    #라벨 크기 조절
    labelcex = 1
)
```
