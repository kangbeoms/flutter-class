### 선굵기 변경

```r
par(bg = 'white')
x <- c(1,3,5,7,9,10)
plot(x,
     # 선과 점이 같이나오는 타입
     type = 'b',
     col = 'red',
     lwd = c(rep(3,1),rep(5,3)),
)
```

```r
# 한 화면에 차트 여러개 그려서 비교하기
# 1에서 100 까지 10개 뽑기
par(bg = 'white')
price <- runif(10,1,100)
plot(price)
```

```r
# type 속성으로 여러화면 그리기
# 2행 2열차트 그리기
par(bg = 'white')
 # 2 행 2열을 그릴거다
par(mfrow = c(2,2))
plot(price, type = 'l')
plot(price, type = 'o')
plot(price, type = 'h')
barplot(price, type = 's')
```

```r
par(bg = 'white')
 # 2 행 2열을 그릴거다
par(mfrow = c(2,2))
plot(price, type = 'o', pch=5)
plot(price, type = 'o', pch=10)
plot(price, type = 'o', pch=15, col ='blue')
plot(price, type = 'o', pch=c(rep(20,8),20,10),
                         cex = c(rep(1,8),3,1))
```

# 수학 함수의 시각화
### y = x^2

```r
f <- function(x) {
    y = x^2
    return(y)
}
```

```r
x <- c(1,2,3)
f(x)
```

```r
par(bg = 'white')
x<- seq(-5, 5, 0.1)
plot(x, f(x), type = 'l')
```

```r
f <- function(x) {
        y <- exp(-x) -4 * x^2 + x + 3 *abs(x) - 3
        return(y)
}
```

```r
par(bg = 'white')
x<- seq(-5, 5, 0.1)
plot(x, f(x), type = 'l')
```

```r
# y= x^2 + 2x +3

f <- function(x) {
        y <- x^2 + 2 * x + 3
        return(y)
}
```

```r
par(bg = 'white')
x<- seq(-5, 5, 0.1)
plot(x, f(x), type = 'l')
```
