## 주성분 분석
: 서로 연관되어 있는 변수들이 관축되었을떄 , 이 변수들이 전체적으로 가지고 있는 정보를 최대학 확보하는 적은수의 새로운 변수를 생성하는 방법

```r
# 설문지를 이용한 주성분 분석
data <- read.table("./Data/pasta.txt")
head(data)
```

```r
# 다중 공선성 확인 : 너무 상관이 높아서 마치 쌍둥이 데이터 인것 찾기
# 너무 같으면 그중 하나만 사용해야한다
cor(data)
```

```r
summary(data)
```

```r
# 표준화 : 평균을 0으로 표준편차를 1로 변경
data1 <- scale(data)
summary(data1)
```

```r
# 표준편차 확인
apply(data1,2,sd)
```

```r
# 주성분 분석 
par(bg = 'white')
p1 <- prcomp(data,scale=T)
biplot(p1) 
```

# USArrests
: 이 데이터세트에는 193년 미국 50개주에서 각각 10만명당 폭행,살인,강간으로 체포된 통계가 있다.
또한 도시지역에 살고있는 인구의 비율도 주어진다.

```r
data(USArrests)
head(USArrests)
tail(USArrests)
```

```r
nrow(USArrests)
```

```r
# 데이터의 상관관계
cor(USArrests)
```

```r
# 주성분 분석
result <- princomp(USArrests, cor = T)
summary(result)
#                    컬럼을 합치는 방법이 4가지가 나옴  
```

```r
par(family = "AppleGothic", bg = 'white') 

biplot(result,
        ) 
```

```r
# Comp1 , Comp2 
result$loadings 
```

제1주성분 = 0.536 * Murder + 0.583 * Assault + 0.278 * UrbanPop +  0.543 * Rape

```r
# 제1주성분으로 새로운 변수값 생성
score <- 0.536 * USArrests$Murder + 0.583 * USArrests$Assault + 0.278 * USArrests$UrbanPop +  0.543 * USArrests$Rape
score
```

```r
# States와 주성분값으로 데이터프레임 생성
Result <- data.frame(States = rownames(USArrests), Score = score)
head(Result)

library(dplyr)

# top 10 뽑기
head(arrange(Result, desc(Score)),10)


```
