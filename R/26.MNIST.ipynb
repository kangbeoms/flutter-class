```r
mnist <- read.csv("./Data/train_1k.csv", header = F)
head(mnist)
```

```r
str(mnist)
```

```r
sqrt(784)
```

```r
# Data 내용 확인하기
num1 = seq(2,785,28)
num2 = seq(29,785, 28)

cnt = 0
for(j in 1:28) {
    n1 = as.numeric(num1[j])
    n2 = as.numeric(num2[j])
    for (i in seq(n1,n2,1)) {
        if (mnist[1,i] != 0) {
            cat("*")
        }else {
            cat(" ")
        }
        # 줄 띄우기
        cnt = cnt +1 
        if (cnt > 27) {
            cat("\n")
            cnt=0
        
        }
    }

}
```

```r
# 정렬하기
library(dplyr)
```

```r
mnist <- arrange(mnist,V1)
head(mnist)
```

```r
tail(mnist)
```

```r
# 번호별 갯수 파악하기
for(i in 0:9) {
    countNum = nrow(subset(mnist,V1 ==i))
    cat(i,":", countNum,"\n")
}
```

```r
# RandomForest는 타겟값이 숫자면 주관식으로 문제를 푼다
mnist$V1 <- as.factor(mnist$V1)
str(mnist)
```

```r
# RandomForest로 예측력과 오분류표 구하기
library(caret)
samp <- createDataPartition(mnist$V1, p=0.7,list = F)

# 학습용
data.tr <- mnist[samp,]
# 검증용
data.test <- mnist[-samp,]

x <- subset(data.test, select = -V1)
y <- data.test$V1

library(randomForest)

# 학습시키기
rf <- randomForest(V1 ~ ., data=data.tr, ntree = 100)
class(rf)

pred <- predict(rf, x)
pred

mean(pred == y)

table(pred,y)
```

```r
t10 <- read.csv("./Data/train_10k.csv")
head(t10)
```

```r
str(t10)
sqrt(784)
table(t10$X5)
```

```r

t10 <- arrange(t10,X5)
head(t10)
tail(t10)
```

```r
t10$X5 <- as.factor(t10$X5)
str(t10)
nrow(t10)
```

```r
# RandomForest 함수
rfFun <- function(A) {
cat("------------------------------------\n")
cat("예측율\n")
    library(caret)
library(randomForest)
for(i in 1:A) {
sampA <- createDataPartition(t10$X5, p=0.7,list = F)
# 학습용
data.trA <- t10[sampA,]
# 검증용
data.testA <- t10[-sampA,]
# 문제
xA <- subset(data.testA, select = -X5)
# 답
yA <- data.testA$X5
# 문제풀기  
rfA <- randomForest(X5 ~ ., data=data.trA, ntree = 100)
# 예측
predA <- predict(rfA, xA)
predA
cat(mean(predA == yA),"\t")
}
cat("\n------------------------------------\n")
cat("\n평균 예측율 :", mean(mean(predA == yA)),"\n")
cat("------------------------------------\n")
}
```
